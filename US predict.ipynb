```python
train = pd.read_csv('train.csv')
train.head()
```

```python
train.drop(['timestamp'], axis=1, inplace= True)
```

```python
train_sample = train.sample(n=20000, random_state=2022)
train_sample
```

```python
pt = train_sample.pivot(index='userId', columns='movieId', values='rating').fillna(0)
pt
```

```python
sim = pd.DataFrame(cosine_similarity(pt), index= pt.index, columns= pt.index)
sim.head()
```

```python
#the ratings given by the most similar users to userId = 94511 of movideId = 3863
most_sim_user_rating = pt.loc[most_sim_user.index, 3863]
```

```python
df = pd.DataFrame(most_sim_user)

df['ratings'] = most_sim_user_rating
df['weighting'] = most_sim_user * df['ratings']
df.rename(columns= {df.columns[0]: 'sim_val'}, inplace=True)


df
```

```python
def avg_movie_rating(movieId):
    col = []
    for movieId in pt.columns:
        for x in pt.loc[:, movieId]:
            if x != 0:
                col.append(x)
        avg = sum(col) / len(col)
        return round(avg, 1)
```

```python
def avg_user_rating(userId):
    col = []
    for userId in pt.index:
        for x in pt.loc[userId, :]:
            if x != 0:
                col.append(x)
        avg = sum(col) / len(col)
        return round(avg, 1)
```

```python
def movie_rater(user_id, movie_id):
    
    
    #case3
    if user_id not in pt.index:
        if movie_id not in pt.columns:
            return train.loc[train['movieId'] == movie_id]['rating'].mean()
   
    #case1
    if user_id not in pt.index:
        if movie_id in pt.columns:
            return avg_movie_rating(movie_id)
    
    #case2
    if movie_id not in pt.columns:
            if user_id in pt.index:
                return avg_user_rating(user_id)
    
    #case0
    if user_id in pt.index:
        if movie_id in pt.columns:
            most_sim_users = sim.loc[:, user_id].drop(user_id).sort_values(ascending=False)[:10]
            most_sim_user_ratings = pt.loc[most_sim_users.index, movie_id]
       
            df1 = pd.DataFrame(most_sim_users) 
            df1['ratings'] = most_sim_user_ratings
            df1['weighting'] = most_sim_users * df1['ratings']
            df1.rename(columns= {df1.columns[0]: 'sim_val'}, inplace=True)
            
            if most_sim_users.sum() == 0:
                return avg_movie_rating(movie_id)
            
            else:            
                pred =  df1['weighting'].sum() / most_sim_users.sum()
                pred1 = (pred*2).round() / 2
                return round(pred1, 1)
                
```

```python
test_combos = []

for x in test.itertuples():
    test_combos.append(x[1:3])
test_combos[:10]

```

```python
test_ratings = []
for idx in test.index: 
    reference_user = test.iloc[idx][0]
    reference_movie = test.iloc[idx][1]
    rating = movie_rater(reference_user, reference_movie)
    test_ratings.append(rating)
```

```python
test_ratings = []
for x, y in test_combos:
    test_ratings.append(movie_rater(x, y))
test_ratings[:10]
```

```python
test['rating'] = test_ratings
test['rating'] = test['rating'].round(decimals=1)
```
